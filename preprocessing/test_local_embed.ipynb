```python
import onnxruntime as ort
print("ORT version:", ort.__version__)
print("Execution providers:", ort.get_available_providers())
```

```python
import onnxruntime as ort, numpy as np
from transformers import AutoTokenizer
import faiss, time

model_dir = "bge-small-en-v1.5-onnx" 
tokenizer = AutoTokenizer.from_pretrained(model_dir)
session   = ort.InferenceSession(f"{model_dir}/onnx/model.onnx",
                                 providers=["CPUExecutionProvider"])

def embed(texts, batch=64):
    chunks, out = [texts[i:i+batch] for i in range(0, len(texts), batch)], []
    for chunk in chunks:
        enc = tokenizer(chunk, padding=True, truncation=True, return_tensors="np")
        (vecs,) = session.run(None, {
            "input_ids": enc["input_ids"],
            "attention_mask": enc["attention_mask"],
        })
        vecs = vecs / np.linalg.norm(vecs, axis=1, keepdims=True)  
        out.append(vecs.astype("float32"))
    return np.vstack(out)

docs  = ["The Eiffel Tower is in Paris."] * 4000
t0    = time.time()
embs  = embed(docs)
print("Encoded 4 k docs in", round(time.time()-t0, 2), "s")

idx   = faiss.IndexFlatIP(embs.shape[1])
idx.add(embs)
```
