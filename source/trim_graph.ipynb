```python
import bz2
import json 
from json import JSONDecodeError

path = "D:/wikipedia/latest-all.json.bz2"

def process(data, buffer, inc):
    i = 0
    lines = []
    while i < len(data):
        if data[i] == '{':
            inc += 1

        if data[i] == '}':
            inc -= 1

        buffer.append(data[i])

        if inc == 0:
            line = ''.join(buffer).strip()
            if line not in [',', '[', ']', '']:
                lines.append(line)
            buffer = []
        
        i += 1

    return lines, buffer, inc

def json_to_edges(line):
    try:
        j = json.loads(line)
        me = j['id']
        claims = j['claims']
        title = j['labels']['en']['value']
    except Exception as e:
        return []
    
    edges = []
    for edge, data in claims.items():
        for datum in data:
            try:
                value = datum['mainsnak']['datavalue']['value']
                if type(value) == dict and 'id' in value:
                    edges.append({
                        'title': title,
                        'me': me,
                        'edge': edge,
                        'neighbor': value['id']
                    })
            except Exception as e:
                pass
    return edges

data = []
buffer = []
n_lines = 0
n_edges = 0
inc = 0
with bz2.BZ2File(path, 'rb') as bz2_file:
    while chunk := bz2_file.read(10000000):
        lines, buffer, inc = process(chunk.decode('utf-8'), buffer, inc)

        n_lines += len(lines)
        with open('data/edges.jsonl', 'a', encoding='utf-8') as file:
            for line in lines:
                edges = json_to_edges(line)
                n_edges += len(edges)
                for edge in edges:
                    file.write(json.dumps(edge) + '\n')
        print(f'Wrote buffer. Total lines: {n_lines}. Total edges: {n_edges}')

```
